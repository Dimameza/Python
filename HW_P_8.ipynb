```python
df1=data[['Total phenols','Flavanoids']]
df2=data[['Total phenols','Proanthocyanins']]
df3=data[['Flavanoids','Proanthocyanins']]
df1.head()
```

```python
df1.plot()
```

```python
corr_s = data[['Total phenols', 'Flavanoids', 'Proanthocyanins']].corr()
sns.heatmap(corr_s, annot=True,xticklabels=corr_s.columns.values,fmt=".1f",yticklabels=corr_s.columns.values,linewidths=0.5)
```

```python
sns.jointplot(x='Total phenols', y='Flavanoids',data=df1, kind='reg');
```

```python
sns.jointplot(x='Total phenols', y='Proanthocyanins',data=df2, kind='reg');
```

```python
sns.jointplot(x='Flavanoids', y='Proanthocyanins',data=df3, kind='reg');
```

```python
# c svg pairplot браузер начинает тормозить
%config InlineBackend.figure_format = 'png' 
sns_plot = sns.pairplot(
    data[['Total phenols', 'Flavanoids', 'Proanthocyanins']]);
sns_plot.savefig('pairplot.png')
```

```python
t1=data['Total phenols']
t2=data['Flavanoids']
stats.ttest_ind(t1, t2)
```

```python
t1=data['Total phenols']
t2=data['Proanthocyanins']
stats.ttest_ind(t1, t2)
```

```python
stats.ttest_rel(t1, t2)
```
